### 1. Binarias eclipsantes

Cuando un sistema de binarias es eclipsante es posible medir con más facilidad los parámetros fundamentales de cada estrella, incluso con telescopios pequeños. De eso, se pueden derivar cantidades de interés como magnitudes, edades y distancias (con mejor precisión que con el método de candelas). El desarrollo de ramas de evolución estelar y astrosismología, entre otras, se ve beneficiado al conseguir parámetros estelares mejores.

[Stanisław et al. 2016](https://arxiv.org/abs/1602.01238)

### 2. Espectrógrafo echelle

Un espectrógrafo echelle utiliza un grating echelle como elemento dispersor. Este tipo de grating tiene menor densidad de ranuras que uno estándar, pero sus ranuras tienen forma de dientes de sierra y están optimizadas para trabajar con ángulos de incidencia mayores. Esto permite su uso para órdenes de difracción mayores, lo que implica mayor dispersión al mismo tiempo que se cubre un buen rango de longitudes de onda. Incluso los espectrógrafos echelle de bajo costo ofrecen resoluciones razonables (R ~ 20 000 en el caso de [PUCHEROS](https://ui.adsabs.harvard.edu/#abs/2012MNRAS.424.2770V/abstract) y [BACHES](https://arxiv.org/abs/1406.2492)) si se utilizan en conjunto con uno o dos prismas u otro dispersor anterior al echelle. En el trabajo de Stanisław se realizan estudios con BACHES y un telescopio de 50cm, comparable al sistema ESO50/PUCHEROS. Ellos concluyen que el sistema es competente para el estudio de binarias eclipsantes.

Podemos comparar esto con lo que ofrecen [FORS](http://www.eso.org/sci/facilities/paranal/instruments/fors/overview.html): R = 2100 en el modo de 456nm-586nm y [HARPS](https://www.eso.org/sci/facilities/lasilla/instruments/harps/inst/description.html): R = 115 000 en 378nm-691nm

Una alta resolución es de especial significado en el estudio de binarias eclipsantes, dado que podría ser posible distinguir el espectro de ambas estrellas.

### 4. Proyección de visibilidad

Se generan gráficos para los candidatos a binarias eclipsante y también para la luna. Se comienza a las 21:00 (hora local)

```python
import ephem
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oduc = ephem.Observer()
oduc.lon = ephem.degrees('289.4656')
oduc.lat = ephem.degrees('-33.2692')
oduc.elevation = 1450

def plotAlt(name,ra,dec):
    s = ephem.FixedBody()
    s._ra = ephem.degrees(ra)
    s._dec = ephem.degrees(dec)

    fig, ax = plt.subplots(3,3,figsize=(15,15))
    x = np.arange(0,4,.5)
    for i in range(3): #recorre filas (semanas)
        oduc.date = '2016/10/29 23:59'
        oduc.date += ephem.hour * 7 * 24 * i
        for j in range(3): #recorre columnas (dias)
            if (j>0):
                oduc.date -= ephem.hour * 4
            oduc.date += ephem.hour * 48
            
            ax[i,j].set_title(name + ' ' + repr(oduc.date.tuple()[0:3]))
            ax[i,j].set_xlabel('Hours since 00:00 UTC')
            ax[i,j].set_ylabel('Altitude (degrees)')
            
            for k in range (8): #recorre cuadros (horas)                
                oduc.date += ephem.minute * 30
                s.compute(oduc)
                
                ax[i,j].plot(x[k], np.degrees(s.alt), 'k.', ms=9)        
    plt.show()
    plt.close()

stars = [['HD23765', '57.1205', '21.7975'],
         ['HD287039', '71.5819', '12.7448'],
         ['HD89849', '155.541', '6.21829'],
         ['BD-07 3648', '203.727', '-8.44247'],
         ['HD198044', '312.091', '-22.7407'],
         ['HD205403', '323.766', '-3.7349'],
         ['HD222891', '356.162', '-8.84879']]

for star in stars:
    plotAlt(star[0],star[1],star[2])
    print 64*'#'
    print 64*'#'
```

### 5. Fechas de cuadratura

Se calculan las cuadraturas a partir del 1 de noviembre del 2016 (MJD 57692). Para ello, primero se calcula la cantidad de eclipses que han ocurrido desde la fecha publicada hasta el 31 de octubre (digamos "hoy"). De esta forma se obtiene la fecha del medio eclipse más cercano después de hoy. Para calcular los siguientes medios eclipses basta mover "i" veces la estrella una cantidad de días de la mitad del período. Para calcular la fecha de cuadratura solamente queda sumar un cuarto de período a la fecha de medio eclipse.

Las tres semanas de cálculo corresponden a los días desde MJD 57692 hasta MJD 57713.

Los gráficos se comienzan en el primer semieclipse correspondiente. Se marcan las cuadraturas que ocurren durante las 4 primeras horas de la noche con una línea verde ancha y las demás con una línea roja.

```python
stars = [['HD198044', 54450.509489, 5.15195],
         ['HD205403', 54123.634818, 2.44490],
         ['HD222891', 54150.650434, 1.59495]]

def quadrature(name,mjd,p):
    n = (57692-mjd)//p # cantidad de eclipses desde el mjd hasta hoy
    now = n*p+mjd # fecha del medio eclipse mas cercano a hoy
    
    x = np.linspace(0,21,500)
    y = np.sin(2*np.pi*x /p)
    fig, ax = plt.subplots(1,1,figsize=(15,5))
    ax.plot(x,y,'k-')
    ax.set_title(star[0])
    ax.set_xlabel('Days from mid-eclipse (MJD ' + repr(now) + ')')
    ax.set_ylabel('Normalized radial velocity')
    ax.set_xlim(0,21)
    for i in range(25):
        quad = now + i*p/2 + p/4
        if(abs(quad) - abs(int(quad))<0.167): # la parte decimal
            plt.axvline(quad-now,color='green',linewidth=4)
        else:
            plt.axvline(quad-now,color='red')
        print quad
    plt.show()
    plt.close()
          
for star in stars:
    print star[0] + '\n'
    quadrature(star[0],star[1],star[2])
    print 64*'#'
```

4 horas corresponden a 0.167 días. Si marcamos el inicio de la noche a las 21:00 GMT-3 coincidimos con las 0:00UTC, luego las cuadraturas que ocurren durante las primeras 4 horas de la noche son las que tienen parte decimal menor a 0.167. Se tabulan las fechas que cumplen el requisito:

### 6. Función de masas

La función de masas para binarias está dada por

$$
f = \dfrac{M_2^3 \sin^3{i}}{(M_1 + M_2)^2} = \dfrac{p v_1^3 \sin^3{i}}{2\pi G}
$$

Como tenemos binarias eclipsantes podemos asumir que el ángulo de inclinación es de 90º. La ecuación queda

$$
\dfrac{M_2^3}{(M_1 + M_2)^2} = \dfrac{p v_1^3}{2\pi G}
$$

Despejando para la velocidad,

$$
v_1 = M_2 \left( \dfrac{2\pi G}{p(M_1 + M_2)^2} \right) ^{(1/3)}
$$

Asumiremos la masa de la estrella principal según su tipo espectral, de acuerdo a los datos en http://www.uni.edu/morgans/astro/course/Notes/section2/spectralmasses.html. 

Si las estrellas compañeras fueran de secuancia principal, podemos asumir alguna relación radio-masa, como, por ejemplo, la que presentan [Gregory A. et al (2012)](https://arxiv.org/abs/1207.3090). Para estrellas de 0.35 radios solares podemos asumir una masa de 0.4 masas solares. Se marca con un punto rojo este caso.

```python
stars = [['HD198044', 1.12, 5.15195],
         ['HD205403', 1.33, 2.44490],
         ['HD222891', 1.10, 1.59495]]

G = 6.67*10**(-11) # gravitacion
M = 1.9891*10**(30) # masa del sol
x = np.linspace(0,2,300)

def plotRV(name,m,p):    
    y = x*M*( (2*np.pi*G) / ((p*24*3600)*(m*M+x*M)**2) )**(0.3333)
    fig, ax = plt.subplots(1,1,figsize=(5,5))
    ax.set_title(star[0])
    ax.set_xlabel('$M_2 (M_{\odot})$')
    ax.set_ylabel('$v_1 (km/s)$')
    ax.plot(x,y/1000,'k-')
    m0 = 0.4
    y0 = m0*M*( (2*np.pi*G) / ((p*24*3600)*(m*M+m0*M)**2) )**(0.3333)
    ax.plot(m0,y0/1000,'r.',ms=9)
    plt.show()
    plt.close()
    print 'velocidad radial esperada: ' + repr(y0/1000) + ' km/s'
    
for star in stars:
    plotRV(star[0],star[1],star[2])
    
       
```

Se esperan velocidades radiales de no menos de 37 km/s. PUCHEROS tiene una resolución promedio de 18 000 y se requiere una de $c/v_1$. Para una velocidad de 37 km/s se requiere $R\approx 8 000$, por lo que no debiera haber problemas en cuanto a resolución.

### 7. Tiempos de exposición

Para calcular el tiempo de exposición podemos asumir que la señal $S \propto t \cdot 10^{-0.4V}$, con $t$ el tiempo de exposición. También podemos asumir que el ruido $N = \sqrt{S}$. Luego,
$$SNR \propto \sqrt{t} \times 10^{-0.2V}$$

La precisión en la velocidad radial $P$ es aproximadamente inversamente proporcional a la SNR, y se sabe que con $V = 3.5$, $P = 100m/s$

$$ 10m/s =  Cte \cdot (\sqrt{300} \times 10^{-0.2\cdot 3.5})^{-1}$$

Obtenemos el valor de la constante

```python
print 0.1/np.sqrt(300)/10**(-0.2*3.5)
```

Para una resolución requerida 10 veces mayor, entonces el tiempo de exposición (en segundos) necesario será:
$$ t = (2.8936 \cdot 10^{-0.2 \cdot V})^{-2} $$

```python
stars = [['HD198044', 7.2],
         ['HD205403', 8.02],
         ['HD222891', 8.07]]

def expTime(V):
    t = (2.8936*10**(-0.2*V))**(-2)
    return t

for star in stars:
    print star[0], expTime(star[1])
```

Como los tiempos son relativamente cortos, se procede a otorgar prioridad según otros factores. Para HD205403 coincidiremos con una cuadratura durante las primera hora de la noche del día 14 de noviembre, asi que se pone en primer lugar. HD222891 es el que se mantiene más alto en el cielo por más tiempo, por lo que se le coloca en último y HD198044 quedaría en segundo puesto.
