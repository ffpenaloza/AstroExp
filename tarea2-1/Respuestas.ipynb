<a href="http://kepler.nasa.gov/Mission/QuickGuide/">Kepler</a> es la décima misión del programa Discovery de la NASA. Fue concebida específicamente para obervar nuestra región en la galaxia en busca de planetas y saber qué tan frecuentes son aquellos similares a la Tierra.
Los objetivos de Kepler son
<ul>
<li>Determinar la fracción de planetas que se encuentran en la zona habitable de su sistema</li>
<li>Determinar la distribución de tamaños y formas de las órbitas planetarias</li>
<li>Estimar la cantidad de planetas en sistemas estelares múltiples</li>
<li>Para los planetas gigantes, determinar reflectividad, tamaños, masas y densidades</li>
<li>Detectar otros integrantes de un sistema estelar una vez descubierto uno, mediante otras técnicas</li>
<li>Conocer las propiedades de las estrellas que tienen un sistema planetario</li>
</ul>

<strong>KIC</strong>, el <em>Kepler Input Catalog</em> es una lista pública de unas 150 000 estrellas de las cuales alrededor de la tercera parte logra ser observada. Cuando se tienen sospechas firmes de que una de ellas alberga al menos un planeta, pasa a ser un <strong>KOI</strong>, <em>Kepler Object of Interest</em>.

En 2013 Kepler Concluyó sus actividades luego de que el segundo de sus tres motores de fallara, impidiendo que pudiera ser apuntado. Repararlo no es una opción ya que se encuentra demasiado lejos de la Tierra (Kepler orbita el sol).

Para continuar la misión se desarrolló una nueva, <em><a href="http://www.nasa.gov/kepler/keplers-second-light-how-k2-will-work">Kepler's Second Light</a></em> <strong>(K2)</strong>, destinada a buscar planetas en el plano de la eclíptica. Sus observaciones se nutren de <strong>EPIC</strong>, <em>Eclpitic Plane Input Catalog</em>.

Según <a href="http://www.nasa.gov/mission_pages/kepler/main/index.html">informa la NASA</a>, de los 4693 candidatos de Kepler 2330 han sido confirmados mientras que de los 458 candidatos de K2 se han confirmado 129.

Se obtienen y se grafican las curvas de luz para cada objeto. La diferencia entre <strong>SAP</strong>, <em>Simple Aperture Photometry</em>, y <strong>PDCSAP</strong>, <em>Pre-search Data Conditioning Simple Aperture Photometry</em>, es que esta última corrige los errores sistemáticos de las series de tiempo SAP. El algoritmo de PDCSAP se encuentra en continuo desarrollo.

```python
import numpy as np
import matplotlib.pyplot as plt
import kplr
%matplotlib inline

client = kplr.API()

star = [client.star(2305930), client.star(4348593), client.star(9469165), client.star(6100702)]

# Se inicializan variables. Las variables en mayusculas contendran todos los datos
lcs, TIME, FLUX, FERR = [], [], [], []

print 'Objeto\t\t\tT eff\t\tRadio\t\tFe/H\t\tLog g'
for s in star:
    # Para la RRab se cambia la cadencia. Produce algo raro en la curva pero sirve para mas adelante
    if(s==star[3]):
        lcs = s.get_light_curves(short_cadence=True)
    else:
        lcs = s.get_light_curves(short_cadence=False)
    time, flux, ferr = [], [], []
    for lc in lcs:
        with lc.open() as f:
            # The lightcurve data are in the first FITS HDU.
            hdu_data = f[1].data
            time.append(hdu_data["time"][~np.isnan(hdu_data["pdcsap_flux"])])
            flux.append(hdu_data["pdcsap_flux"][~np.isnan(hdu_data["pdcsap_flux"])])
            ferr.append(hdu_data["pdcsap_flux_err"][~np.isnan(hdu_data["pdcsap_flux"])])
            # Ignora los NaN al hacer append
    TIME.append(time)
    FLUX.append(flux)
    FERR.append(ferr)
    print 'KIC ' + repr(s.kic_kepler_id) + '\t\t' + repr(s.kic_teff) + '\t\t' + repr(s.kic_radius) + '\t\t' + repr(s.kic_feh) + '\t\t' + repr(s.kic_logg)

fig, ax = plt.subplots(4, 1,figsize=(15,25))
       
for i in range(0,4):    
    ax[i].set_title('KIC ' + repr(star[i].kic_kepler_id),size=16)
    ax[i].set_xlabel('Time',size=14)
    ax[i].set_ylabel('Flux',size=14)
    time = TIME[i]
    flux = FLUX[i]
    for j in range(0,len(time)):
        ax[i].plot(time[j],flux[j],'k.',alpha=.1)

plt.show()
plt.close()
```

Por comodidad se incluye de inmediato la curva del objeto de la pregunta 3. El objeto escogido es el <a href="https://arxiv.org/pdf/1108.5683.pdf">KIC 6100702</a>, de período 0.4881457 días.

<h2>Pregunta 6</h2>

<a href="http://exoplanetarchive.ipac.caltech.edu/applications/Periodogram/docs/Algorithms.html">Lomb-Scargle</a> es un algoritmo, variación de la transformada de Fourier discreta, que consiste en ajustar combinaciones de funciones sinusoidales a una serie de tiempo. Como resultado se obtiene un periodograma (de Lomb-Scargle).

```python
from gatspy import periodic
import warnings

def LombSPeriod(time,flux,ferr):
    model = periodic.LombScargleFast().fit(time, flux, ferr)
    model.optimizer.period_range=(5, 55)
    period.append(model.best_period)
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        scores = model.score(periods)
    predict.append(model.predict(periods, period=model.best_period))
    return period, scores, predict

fig, ax = plt.subplots(3, 1,figsize=(15,25))
periods = np.linspace(0.1, 140, 1000)
period, predict = [], []
for i in range(0,3):
    ax[i].set_title('KIC ' + repr(star[i].kic_kepler_id),size=16)
    ax[i].set_xlabel('Period',size=14)
    ax[i].set_ylabel('Lomb-Scargle Power',size=14)
    time = np.concatenate(tuple(TIME[i]))
    flux = np.concatenate(tuple(FLUX[i]))
    ferr = np.concatenate(tuple(FERR[i]))
    period, scores, predict = LombSPeriod(time,flux,ferr)
    ax[i].plot(periods, scores,'k-')        
    print '\n\tKIC ' + repr(star[i].kic_kepler_id) + ': Período = ' + repr(period[i]) + '\n'
    
plt.show()
plt.close()
```

Veamos qué pasa si se grafican las fases con los períodos obtenidos

Existe, por lo tanto, variabilidad. Además, de las curvas entregadas por el modelo Lomb-Scargle, puede verse que existe cierta periodicidad si se grafica la fase. <h2>Pregunta 7</h2>

<a href="https://www.mdscc.nasa.gov/index.php?Section=Estrellas_variables">La variabilidad</a> puede deberse tanto a factores extrínsecos como intrínsecos. Factores extrínsecos puede ser, por ejemplo, la obstrucción de la luz que recibimos por culpa de un objeto entre nosotros y la estrella que se observa (un planeta, otra estrella, etc).

Factores intrínsecos pueden ser
<ul>
<li>Pulsaciones, en que la estrella realmente modifica su radio periódicamente</li>
<li>Erupciones irregulares en la superficie</li>
<li>Manchas estelares y su cambio de posición</li>
</ul>

En las estrellas estudiadas si bien se aprecia variación en el flujo y cierta periodicidad, difícilmente son estrellas pulsantes (al menos notoriamente). Los períodos vistos en los gráfico de flujo v/s fase dejan ver que hay varios subperíodos que se superponen, lo que podría deberse a oscilaciones o pulsaciones superpuestas. De todas formas, el período principal (que calcula el algoritmo de Lomb-Scargle) es lo suficientemente extenso como para pensar que la mayor variación se debe simplemente a la rotación de la estrella. Esto, en todo caso, no deja de ser interesante puesto que se trata de estrellas gigantes, lo que supone velocidades superficiales de rotación muy grandes.

Podríamos decir que estas estrellas son variables semi regulares.
